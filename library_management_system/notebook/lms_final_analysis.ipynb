```python
# Merge all Table : Book_Issue,Book_Details,Lms_Member,Suppliers_Details,Fine_Details
lms = Book_Issue.merge(Book_Details, on="BOOK_CODE", how="left").merge(Lms_Members, on="MEMBER_ID", how="left").merge(Suppliers_Details, on="SUPPLIER_ID", how="left").merge(Fine_Details, on="FINE_RANGE", how="left")
lms.head()
```

```python
# Check for missing entries
total = lms.isnull().sum().sort_values(ascending=False)
percent=(lms.isnull().sum()*100/len(lms)).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1,keys=['Total', 'Percent'])
missing_data.head(20)
```

```python
# Check for missing entries
lms['MAX_DAYS_DELAYED'] = lms['MAX_DAYS_DELAYED'].fillna(0)
lms['FINE_AMOUNT'] = lms['FINE_AMOUNT'].fillna(0)
```

##  1. Days Delayed (Return Delay Analysis)

###  Insight:
Some members return books late, but most delays are under 10 days.  
The average delay across the dataset is approximately 6.7 days.  
33% of the records show a delay in return, yet not all late returns result in fines. This may suggest inconsistencies in fine enforcement or rules allowing grace periods.

###  Recommendation:
- Introduce automated reminders via SMS or email 23 days before the return date to reduce late returns.
- Review and enforce consistent fine policies based on the number of delay days.
- Consider displaying delay history for members to increase accountability.

---

##  2. Book Popularity

###  Insight:
The most borrowed books are:
- *Programming in ANSI C*
- *Let Us C*
- *Learning Python*
- *Effective Modern C++*

These top-performing books fall under the C, C++, and Python programming categories, highlighting a strong interest among memberslikely for academic studies, coding practice, or professional skill-building. The consistent borrowing of these titles suggests they are essential resources for many users.


###  Recommendation:
- Consider adding more copies of these high-demand titles to reduce wait times.
- Expand the programming collection by including related topics like JavaScript, Data Structures, Machine Learning, and Web Development.
- Feature these books in a dedicated Most Popular shelf or digital highlight to drive further engagement.

---

##  3. Fine per Member

###  Insight:
A small number of members account for a large percentage of total fines.  
For example, members like *Amruta* and *Ashish* contribute the highest amounts.  
This may suggest habitual late returns by a few users or difficulty accessing return facilities.

###  Recommendation:
- Identify and engage directly with high-fine members to understand why delays occur.
- Offer flexible fine payment options for frequent offenders.
- Implement a member points or trust system where consistent early returners are rewarded with extended borrowing time.

---

##  4. Fine per City

###  Insight:
CHENNAI has the highest total fines compared to other cities.  
This may correlate with either a larger membership base in Chennai or operational inefficiencies in that location.

###  Recommendation:
- Conduct a location-based audit to understand why fines are concentrated in Chennai.
- Provide staff training or better systems in high-fine areas to streamline the return and reminder processes.
- Consider location-specific campaigns for timely book returns.

---

##  5. Fine per Book Category

###  Insight:
Books in the **C PROGRAMMING** and **PYTHON PROGRAMMING** categories contribute the highest fine amounts.  
This correlates with their high popularity but also indicates they are more likely to be returned late.

###  Recommendation:
- Review loan durations for high-fine categories; shorter durations might not match actual usage needs.
- Introduce tailored return policies or reminders based on the book category.
- Consider increasing copies of these books to reduce waiting queues that may lead to delayed returns.

---

##  6. Average Fine by Membership Status

###  Insight:
Permanent members incur more fines on average than temporary members.  
This could be due to less familiarity with the system or fewer privileges (e.g., shorter borrowing windows).

###  Recommendation:
- Educate permanent members on borrowing and return policies at the time of registration.
- Encourage conversion to temporary membership with loyalty rewards or additional benefits.
- Consider stricter borrowing limits or shorter durations for permanent users.

---

##  7. Supplier Performance (Total Book Value)

###  Insight:
Suppliers like **SINGAPORE SHOPPEE** and **ROSE BOOK STORE** provide books with the highest total value.  
These suppliers contribute significantly to the collection and likely supply in-demand, high-quality titles.

###  Recommendation:
- Strengthen partnerships with top-performing suppliers through volume contracts or priority procurement.
- Monitor the delivery efficiency and cost-effectiveness of each supplier periodically.
- Consider rewarding reliable suppliers with more frequent orders or referrals.

---



##  Final Summary Insight  
 **Key Takeaway:**  
The data clearly shows that programming books (especially C, C++, and Python) are the most popular and most borrowed, indicating a strong interest in tech-related content  likely from students or professionals. While most members return books on time, a third return them late, and not all late returns lead to fines. This suggests there may be some inconsistency in how fines are applied.

Fines are especially high among a few members and in certain cities like Chennai. Meanwhile, permanent members tend to accumulate more fines than temporary ones, possibly due to borrowing more frequently or longer periods.

Suppliers like **Singapore Shoppee** and **Rose Book Store** provide the highest-value books, making them key contributors to the librarys collection.

Overall, the insights show that improving book availability, sending timely reminders, and standardizing return/fine policies could significantly enhance the librarys efficiency and member experience.


##  Final Recommendations

Based on the analysis, here are a few key suggestions to help improve the librarys efficiency and user experience:

-  Add more copies of the most popular programming books like C, C++, and Python titles  theyre clearly in high demand and get borrowed often.
-  Set up simple return reminders (via SMS or email) to reduce late returns and help members stay on track.
-  Take a closer look at why fines are highest in Chennai  there might be issues with processes or communication in that location.
-  Support members who often return books late  it could be due to access issues or lack of awareness about rules.
-  Keep strong relationships with reliable suppliers like Singapore Shoppee and Rose Book Store  they provide valuable books that people actually use.

These small improvements can go a long way in making the library more helpful, fair, and user-friendly for everyone.
